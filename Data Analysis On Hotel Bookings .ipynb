# Data Analysis and Visualization

```python
# Cancellation percentage
can_per = df['is_canceled'].value_counts(normalize = True)*100
print(can_per)
```

```python
plt.figure(figsize = (8,6))
plt.title('Reservation Status Count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor = 'k',width = 0.5)
plt.savefig('Reservation Status Count.png')
plt.show()
```

```python
# Cancellation percentage hotel type wise....
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.savefig('Reservation status in different hotels.png')
plt.show()
```

```python
# Cancelation Percentage Hotel Type Wise
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)*100
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)*100
```

```python
# Calculating the average daily rate for City and Resort Hotel
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.savefig('Average Daily Rate in City and Resort Hotel.png')
plt.show()
```

```python
# Month wise reservation status
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.savefig('Reservation status per month.png')
plt.show()
```

```python
# Average Daily Rate for each month for only canceled Bookings
plt.figure(figsize = (15,8))
plt.title('ADR Per Month', fontsize = 30)
adr_per_month = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x = 'month', y = 'adr', data = adr_per_month)
plt.savefig('ADR Per Month.png')
plt.show()
```

```python
# Cancelation rate based on countries
cancel_data = df[df['is_canceled'] == 1]
top_10_country = cancel_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 Countries with Higest Cancelation Rate')
plt.pie(top_10_country, autopct = '%.2f',labels = top_10_country.index)
plt.savefig('Top 10 Countries with Higest Cancelation Rate.png')
plt.show()
```

```python
# Average Daily Rate for canceled and not canceled bookings
```

```python
cancelled_df_adr = cancel_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.savefig('Average Daily Rate.png')
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```
